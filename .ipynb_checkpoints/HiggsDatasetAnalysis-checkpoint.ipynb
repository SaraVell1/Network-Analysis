```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy
from networkx import hits
from networkx import pagerank
from networkx import eigenvector_centrality
from networkx import load_centrality

G =nx.read_edgelist('Dataset/higgs-activity_timeRed.txt', create_using = nx.Graph(), nodetype=int)
print(G)
pos = nx.circular_layout(G)
plt.figure(figsize=(10,10))
nx.draw_networkx(G, pos, node_size=10, with_labels=False, node_color='purple', node_shape='*')
plt.show()
```

```python
number_of_nodes = nx.number_of_nodes(G)
number_of_edges = nx.number_of_edges(G)
totalTriangles = nx.triangles(G)
number_of_triangles = sum(totalTriangles.values()) / 3
print(number_of_nodes)
print(number_of_edges)
print(number_of_triangles)
```

```python
dCen= nx.degree_centrality(G)
```

```python
node_color = [20000.0 * G.degree(v) for v in G]
node_size = [v * 10000 for v in dCen.values()]
plt.figure(figsize=(10,10))
nx.draw_networkx(G, pos=pos, with_labels=False,
node_color=node_color,
node_size=node_size)
plt.axis('off')
#sorted(dCen, reverse=True)[:5]
```

```python
myK = 10
betCent = nx.betweenness_centrality(G, k= myK)
```

```python
node_color = [20000.0 * G.degree(v) for v in G]
node_size = [v * 10000 for v in betCent.values()]
plt.figure(figsize=(10,10))
nx.draw_networkx(G, pos=pos, with_labels=False,
node_color=node_color,
node_size=node_size)
plt.axis('off')
sorted(betCent, key=betCent.get, reverse=True)[:5]
```

```python
k = 88
closenessCent = nx.closeness_centrality(G, u=k, distance=None, wf_improved=True)
node_color = [20000.0 * G.degree(v) for v in G]
if k == 88: node_size== k*10000 
else: node_size = 10
plt.figure(figsize=(10,10))
nx.draw_networkx(G, pos=pos, with_labels=False,
node_color=node_color, node_size= node_size)
plt.axis('off')
```

```python
EigCent= eigenvector_centrality(G, max_iter=300, tol=1e-03, nstart=None, weight=None)
```

```python
node_color = [2000.0 * G.degree(v) for v in G]
node_size = [v * 10000 for v in EigCent.values()]
plt.figure(figsize=(10,10))
nx.draw_networkx(G, pos=pos, with_labels=False,
node_color=node_color,
node_size=node_size)
sorted(EigCent, key=EigCent.get, reverse=True)[:5]
```
